```python
#Distribution of Globalsales
plt.figure(figsize=(12,6))
sns.histplot(global_sales, bins=20,kde=True)
plt.title('Distribution of Global Sales')
plt.xlabel('Global Sales')
plt.ylabel('Frequency')
plt.show()
```

```python
#most popualar Genres
top_genres=df['Genre'].value_counts().head(10)
plt.figure(figsize=(10,6))
sns.barplot(x=top_genres.index, y=top_genres.values)
plt.title('Top 10 Popular Genres')
plt.xlabel('Genre')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
#most popualar Platforms
top_Platform=df ['Platform'].value_counts().head(10)
plt.figure(figsize=(10,6))
sns.barplot(x=top_Platform.index, y=top_Platform.values)
plt.title('Top 10 Popular Platform')
plt.xlabel('Platform')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
#most popualar Publishers
top_Publisher=df['Publisher'].value_counts().head(10)
plt.figure(figsize=(10,6))
sns.barplot(x=top_Publisher.index, y=top_Publisher.values)
plt.title('Top 10 Popular Publisher')
plt.xlabel('Publisher')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```
